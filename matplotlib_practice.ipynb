```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

z_axis = [12, 15, 14, 17, 42, 36, 31, 25, 28, 24, 36, 34]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.xlim("Jan", "May")
plt.grid()
plt.xlabel("month")
plt.ylabel("total US dollars")
plt.plot(x_axis, z_axis, label = "new")
#plt.figure(figsize = (15, 15))
```

```python
plt.subplots()
```

```python
# Create the plot with ax.plt()
plt.figure(figsize = (15, 15))
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = "Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker = "o", color = "blue", linewidth = 3, label = "Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
(fig, ax) = plt.subplots()
ax.plot(x_axis, y_axis)

# Change the line color to green
ax.plot(x_axis, y_axis, color = "green", marker = "D", linewidth = 2, label = "Boston")
ax.set_ylim(0, 45)
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
ax.grid()
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, "o")
```

```python
plt.scatter(y_axis, x_axis, color = "red", label = "Chicago", s = y_axis)
plt.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
(fig, ax) = plt.subplots()
ax.scatter(y_axis, x_axis, color = "blue", s = [5 * i for i in y_axis], alpha = 0.5, edgecolors = "black", linewidths = 2, label = "Boston" )
ax.legend()
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.set_xlim(0, 50)
```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
plt.subplots(figsize = (8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels = x_axis, explode = explode_values, colors = colors, autopct = "%.1f%%")

plt.show()
```
